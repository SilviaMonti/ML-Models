```python
#we don't want categorical variables, so if we have some, we need to transform them into dummy variables 
cat_feats = ['purpose']
final_data = pd.get_dummies(loans,columns=cat_feats,drop_first=True)
```

```python
#now we want to predict if the loan will be paid or not
from sklearn.model_selection import train_test_split

X = final_data.drop('not.fully.paid',axis=1)
y = final_data['not.fully.paid']

X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=0.30,random_state=101)
```

```python
#now let's build the decision tree
from sklearn.tree import DecisionTreeClassifier

dtree = DecisionTreeClassifier()
dtree.fit(X_train,y_train)
```

```python
predictions = dtree.predict(X_test)
```

```python
from sklearn.metrics import classification_report,confusion_matrix
print(classification_report(y_test,predictions))
```

```python
print(confusion_matrix(y_test,predictions))
```

```python
#now let's see the difference with the random forest model
from sklearn.ensemble import RandomForestClassifier
rfc = RandomForestClassifier()
rfc.fit(X_train,y_train)
```

```python
predictions_rfc = rfc.predict(X_test)
```

```python
print(classification_report(y_test,predictions_rfc))
```

```python
print(confusion_matrix(y_test,predictions_rfc))
```
